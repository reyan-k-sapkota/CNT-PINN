```python
import torch
import time

# Settings
batch_size = 32      # number of matrices
matrix_size = 1024   # 1024x1024 matrices → moderate memory
num_iters = 20       

# Function to benchmark a device
def benchmark(device):
    
    a = torch.randn(batch_size, matrix_size, matrix_size, device=device)
    b = torch.randn(batch_size, matrix_size, matrix_size, device=device)

    # Warm-up 
    _ = torch.matmul(a, b)

    # Timed iterations
    start = time.time()
    for _ in range(num_iters):
        c = torch.matmul(a, b)
    if device.type == 'cuda':
        torch.cuda.synchronize()  
    end = time.time()
    return end - start

#  CPU benchmark
cpu_time = benchmark(torch.device("cpu"))
print(f"CPU: {cpu_time:.4f} seconds for {num_iters} batched matmuls")

# GPU benchmark (if available)
if torch.cuda.is_available():
    gpu_time = benchmark(torch.device("cuda:0"))
    print(f"GPU: {gpu_time:.4f} seconds for {num_iters} batched matmuls")
    print(f"GPU used: {torch.cuda.get_device_name(0)}")
else:
    print("GPU not available for benchmarking.")
```
